```python
from google.colab.patches import cv2_imshow
import cv2

image = cv2.imread('/content/hassan2.png')
if image is not None:
    cv2_imshow(image)  # Use cv2_imshow to display the image
else:
    print("Impossible d'afficher l'image")
```

```python
# Load the original image
image = cv2.imread('/content/hassan2.png')

if image is not None:
    # Resize the image to a new width and height
    new_width = 400   # Set the new width you want
    new_height = 300  # Set the new height you want
    resized_image = cv2.resize(image, (new_width, new_height))

    # Display the resized image
    cv2_imshow(resized_image)
else:
    print("Impossible d'afficher l'image")
```

```python
if image is not None:
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)

    # Display the grayscale image
    cv2_imshow(gray_image)
else:
    print("Impossible d'afficher l'image")
```

```python
if resized_image is not None:
    # Apply Gaussian blur to the color image
    blurred_image = cv2.GaussianBlur(resized_image, (5, 5), 0)  # You can adjust the kernel size (e.g., (5, 5)) and standard deviation (0) as needed

    # Display the blurred color image
    cv2_imshow(blurred_image)
else:
    print("Impossible d'afficher l'image")
```

```python
import numpy as np
import matplotlib.pyplot as plt
if resized_image is not None:
    # Calculate the histogram
    hist = cv2.calcHist([resized_image], [0], None, [256], [0, 256])

    # Flatten the histogram into a 1D array
    hist = hist.flatten()

    # Plot the histogram
    plt.figure(figsize=(8, 4))
    plt.title("Histogram")
    plt.xlabel("Pixel Value")
    plt.ylabel("Frequency")
    plt.plot(hist, color='black')
    plt.xlim([0, 256])  # Set the x-axis limits from 0 to 255 (for 8-bit images)

    plt.show()
else:
    print("Impossible de charger l'image")
```

```python
# # Load the image in grayscale mode
# image = cv2.imread('/content/hassan2.png', cv2.IMREAD_GRAYSCALE)

if gray_image is not None:
    # Apply a threshold
    threshold_value = 128  # Adjust this value as needed
    _, thresholded_image = cv2.threshold(gray_image, threshold_value, 255, cv2.THRESH_BINARY)

    # Display the thresholded image
    cv2_imshow(thresholded_image)
else:
    print("Impossible de charger l'image")
```
