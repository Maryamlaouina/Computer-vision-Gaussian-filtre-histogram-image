# Chargement des images

```python
import cv2
from google.colab.patches import cv2_imshow
import os

# Chemin vers le dossier contenant les images
dossier_images = '/content/images'

# Liste des extensions d'images prises en charge
extensions_images = ['.jpg', '.jpeg', '.png']

# Spécifier la nouvelle largeur et hauteur
nouvelle_largeur = 300
nouvelle_hauteur = 200

# Nombre d'images affichées par ligne
images_par_ligne = 3

# Initialiser une liste pour stocker les images à afficher dans chaque ligne
ligne_images = []

# Parcourir les fichiers du dossier
for fichier in os.listdir(dossier_images):
    if fichier.endswith(tuple(extensions_images)):
        # Construire le chemin complet du fichier
        chemin_image = os.path.join(dossier_images, fichier)

        # Charger l'image
        image = cv2.imread(chemin_image)

        # Vérifier si l'image a été chargée correctement
        if image is not None:
            # Redimensionner l'image
            image_redimensionnee = cv2.resize(image, (nouvelle_largeur, nouvelle_hauteur))

            # Ajouter l'image à la liste des images de la ligne actuelle
            ligne_images.append(image_redimensionnee)

            # Si nous avons atteint le nombre d'images par ligne, afficher la ligne et réinitialiser
            if len(ligne_images) == images_par_ligne:
                ligne_affichage = cv2.hconcat(ligne_images)
                cv2_imshow(ligne_affichage)
                ligne_images = []

# Afficher les images restantes dans la dernière ligne
if ligne_images:
    ligne_affichage = cv2.hconcat(ligne_images)
    cv2_imshow(ligne_affichage)
```

# Reduction du bruit avec un filtre gaussien

```python
import cv2
import matplotlib.pyplot as plt

# Charger l'image
image = cv2.imread('/content/vegetables.jpg')

# Spécifier la nouvelle largeur et hauteur
nouvelle_largeur = 500  # Remplacez par la nouvelle largeur souhaitée
nouvelle_hauteur = 300  # Remplacez par la nouvelle hauteur souhaitée

# Vérifier si l'image a été chargée correctement
if image is not None:
    # Redimensionner l'image
    image_redimensionnee = cv2.resize(image, (nouvelle_largeur, nouvelle_hauteur))

    # Appliquer un filtre gaussien à l'image redimensionnée
    image_filtree = cv2.GaussianBlur(image_redimensionnee, (5, 5), 0)

    # Afficher l'image originale, l'image redimensionnée et l'image filtrée
    plt.figure(figsize=(10, 5))
    plt.subplot(131)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title("Image Originale")
    plt.axis('off')

    plt.subplot(132)
    plt.imshow(cv2.cvtColor(image_redimensionnee, cv2.COLOR_BGR2RGB))
    plt.title("Image Redimensionnée")
    plt.axis('off')

    plt.subplot(133)
    plt.imshow(cv2.cvtColor(image_filtree, cv2.COLOR_BGR2RGB))
    plt.title("Image Filtrée")
    plt.axis('off')

    plt.show()
else:
    print("Impossible de charger l'image.")
```

# Filtre Médian

```python
import cv2
import matplotlib.pyplot as plt

# Charger l'image
image = cv2.imread('/content/vegetables.jpg')

# Spécifier la nouvelle largeur et hauteur
nouvelle_largeur = 500  # Remplacez par la nouvelle largeur souhaitée
nouvelle_hauteur = 300  # Remplacez par la nouvelle hauteur souhaitée

# Vérifier si l'image a été chargée correctement
if image is not None:
    # Redimensionner l'image
    image_redimensionnee = cv2.resize(image, (nouvelle_largeur, nouvelle_hauteur))

    # Appliquer un filtre médian à l'image redimensionnée
    image_filtree = cv2.medianBlur(image_redimensionnee, 5)

    # Afficher l'image originale, l'image redimensionnée et l'image filtrée
    plt.figure(figsize=(10, 5))
    plt.subplot(131)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title("Image Originale")
    plt.axis('off')

    plt.subplot(132)
    plt.imshow(cv2.cvtColor(image_redimensionnee, cv2.COLOR_BGR2RGB))
    plt.title("Image Redimensionnée")
    plt.axis('off')

    plt.subplot(133)
    plt.imshow(cv2.cvtColor(image_filtree, cv2.COLOR_BGR2RGB))
    plt.title("Image Filtrée (Filtre Médian)")
    plt.axis('off')

    plt.show()
else:
    print("Impossible de charger l'image.")
```

# Image avec Balance des Blancs Corrigée

```python
import cv2
import matplotlib.pyplot as plt

# Charger l'image
image = cv2.imread('/content/vegetables.jpg')

# Vérifier si l'image a été chargée correctement
if image is not None:
    # Appliquer un ajustement des couleurs pour corriger la balance des blancs
    image_balance_des_blancs = cv2.xphoto.createSimpleWB()
    image_balancee = image_balance_des_blancs.balanceWhite(image)

    # Afficher l'image originale et l'image avec la balance des blancs corrigée
    plt.figure(figsize=(10, 5))
    plt.subplot(121)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title("Image Originale")
    plt.axis('off')

    plt.subplot(122)
    plt.imshow(cv2.cvtColor(image_balancee, cv2.COLOR_BGR2RGB))
    plt.title("Image avec Balance des Blancs Corrigée")
    plt.axis('off')

    plt.show()
else:
    print("Impossible de charger l'image.")
```

# Image avec Correction Gamma

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Charger l'image
image = cv2.imread('/content/vegetables.jpg')

# Vérifier si l'image a été chargée correctement
if image is not None:
    # Appliquer la correction gamma en ajustant la luminance
    gamma = 1.5  # Valeur du paramètre gamma (ajustez selon vos besoins)
    image_corrigee = np.power(image / 255.0, gamma) * 255.0

    # Assurez-vous que les valeurs sont comprises entre 0 et 255
    image_corrigee = np.uint8(image_corrigee)

    # Afficher l'image originale et l'image avec la correction gamma
    plt.figure(figsize=(10, 5))
    plt.subplot(121)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title("Image Originale")
    plt.axis('off')

    plt.subplot(122)
    plt.imshow(cv2.cvtColor(image_corrigee, cv2.COLOR_BGR2RGB))
    plt.title("Image avec Correction Gamma")
    plt.axis('off')

    plt.show()
else:
    print("Impossible de charger l'image.")
```

# Image avec Détails Améliorés (Opérateur Laplacien)

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Charger l'image
image = cv2.imread('/content/vegetables.jpg')

# Vérifier si l'image a été chargée correctement
if image is not None:
    # Convertir l'image en niveaux de gris
    image_grise = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Appliquer l'opérateur Laplacien
    image_laplacien = cv2.Laplacian(image_grise, cv2.CV_64F)

    # Assurez-vous que les valeurs sont comprises entre 0 et 255
    image_laplacien = cv2.convertScaleAbs(image_laplacien)

    # Afficher l'image originale et l'image avec les détails améliorés
    plt.figure(figsize=(10, 5))
    plt.subplot(121)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title("Image Originale")
    plt.axis('off')

    plt.subplot(122)
    plt.imshow(image_laplacien, cmap='gray')
    plt.title("Image avec Détails Améliorés (Opérateur Laplacien)")
    plt.axis('off')

    plt.show()
else:
    print("Impossible de charger l'image.")
```
